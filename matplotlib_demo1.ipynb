```python
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# Phần nâng cao hơn với matplotlib có thể bao gồm:
## - Tạo các biểu đồ phức tạp hơn như heatmaps, 3D plots, contour plots.
## - Sử dụng các thư viện bổ sung như Seaborn để tạo các biểu đồ thống kê đẹp mắt hơn.
## - Tích hợp matplotlib với các công cụ khác như Pandas để trực quan hóa dữ liệu từ DataFrame.
## - Tạo các dashboard tương tác với matplotlib và các thư viện như Plotly hoặc Bokeh.
## - Tùy chỉnh sâu hơn về mặt thẩm mỹ và chức năng của biểu đồ.
## - Tạo các biểu đồ động hoặc tương tác sử dụng matplotlib animation hoặc mpld3.
## - Xuất biểu đồ sang các định dạng khác nhau như SVG, PDF với chất lượng cao.
## - Sử dụng các colormaps tùy chỉnh để làm nổi bật dữ liệu.
## - Tạo các subplot phức tạp với nhiều hàng và cột, chia sẻ trục và tùy chỉnh từng subplot riêng biệt.
## - Tích hợp matplotlib với các GUI frameworks như Tkinter, PyQt để tạo ứng dụng desktop có biểu đồ.
## - Sử dụng các công cụ như Jupyter Notebook để tạo các báo cáo tương tác với biểu đồ matplotlib.

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 3, 5, 7, 11]
plt.stackplot(x, y1, y2, labels=['y=x^2', 'y=prime'], colors=['lightblue', 'lightgreen'])
plt.title('Stacked Area Chart Example')
plt.legend(loc='upper left')
plt.show()
```

```python
years = [2018, 2019, 2020, 2021, 2022]
noi = [200, 220, 250, 270, 300]
ngoai = [150, 160, 180, 210, 230]
nhi = [100, 120, 150, 180, 190]

plt.figure(figsize=(10, 6))
plt.plot(years, noi, 'ro--', label='Nội')
plt.plot(years, ngoai, 'go--', label='Ngoại')
plt.plot(years, nhi, 'bo--', label='Nhi')
plt.title('Số ca bệnh theo từng khoa mỗi năm')
plt.xlabel('Năm')
plt.ylabel('Số ca bệnh')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.stackplot(years, noi, ngoai, nhi, labels=['Nội', 'Ngoại', 'Nhi'], colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Số ca bệnh theo từng khoa mỗi năm')
plt.xlabel('Năm')
plt.ylabel('Số ca bệnh')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [3, 7 , 5]
plt.barh(categories, values, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Horizontal Bar Chart Example')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
# huyết áp
noi = [120, 125, 130, 135, 140, 145, 150, 160, 170]
ngoai = [110, 115, 120, 125, 128, 130, 135, 140, 145, 150]
nhi = [90, 95, 98, 100, 102, 105, 110, 115, 120, 125]

plt.figure(figsize=(10, 6))
plt.boxplot([noi, ngoai, nhi], tick_labels=['Nội', 'Ngoại', 'Nhi'])
plt.title('Box Plot of Blood Pressure by Department')
plt.ylabel('Blood Pressure (mmHg)')
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(10, 6))
plt.violinplot(data, showmeans=True)
plt.title('Violin Plot Example')
plt.show()
```

```python
# Sử dụng subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.plot([1, 2, 3], [1, 4, 9], 'ro--')
ax1.set_title('Line Plot')
ax2.bar(['A', 'B', 'C'], [3, 7, 5], color='lightblue')
ax2.set_title('Bar Chart')
plt.show()
```

```python
plt.plot([1, 2, 3], [1, 4, 9], 'ro--') # Vẽ đường biểu diễn y = x^2
plt.annotate("Max", xy=(3, 9), xytext=(2, 8),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12, color='red')
plt.title('Annotated Plot Example')
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]
errors = [0.2, 0.3, 0.1]
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5, linestyle='--', color='blue')
plt.title('Error Bar Example')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
# plt.fill_between(x, y, where=(y > 0), color='lightblue', alpha=0.5, label='Positive Area')
# plt.fill_between(x, y, where=(y <= 0), color='lightcoral', alpha=0.5, label='Negative Area')
plt.fill_between(x, y-0.1, y+0.1, color='lightblue', alpha=0.5)
plt.title('Sine Wave')
# plt.legend()
plt.show()
```

```python
# from scipy.stats import gaussian_kde
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot Example')
plt.show()
```
