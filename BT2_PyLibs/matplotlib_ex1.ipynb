```python
import matplotlib.patches as mpatches

# Đọc file CSV
df = pd.read_csv('HR_comma_sep.csv')

# Đếm số lượng nhân viên theo mức lương
salary_counts = df['salary'].value_counts()

# Vẽ biểu đồ cột Bar Chart với màu sắc khác nhau cho từng cột
plt.figure(figsize=(8, 5))
bars = salary_counts.plot(kind='bar', 
                          color=['red', 'orange', 'green'],  # Mỗi cột một màu
                          edgecolor='black')

# Thêm tiêu đề và nhãn trục
plt.title('Số lượng nhân viên theo mức lương', fontsize=16, fontweight='bold')
plt.xlabel('Mức lương', fontsize=12)
plt.ylabel('Số lượng nhân viên', fontsize=12)

# Điều chỉnh nhãn trục X
plt.xticks(rotation=0, fontsize=11)

# Tạo legend với màu sắc tương ứng
legend_labels = ['Lương thấp', 'Lương trung bình', 'Lương cao']
handles = [mpatches.Patch(color='red', label=legend_labels[0]),
           mpatches.Patch(color='orange', label=legend_labels[1]),
           mpatches.Patch(color='green', label=legend_labels[2])]

# Thêm chú thích cho biểu đồ
plt.legend(handles=handles, loc='upper right', fontsize=12, title='Mức lương')

# Cải thiện cách trình bày biểu đồ
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
import seaborn as sns

# Đọc dữ liệu từ file CSV
Student_Pandas = pd.read_csv('student-mat.csv', sep=';')
# Loại bỏ các dòng có giá trị NaN trong các cột cần vẽ
filtered_data = Student_Pandas[['G1', 'G2', 'G3', 'studytime']].dropna()

# Vẽ lmplot: studytime và G3 với hue='sex'
sns.lmplot(x='studytime', y='G3', data=Student_Pandas, hue='sex', aspect=1.5, height=6)
plt.title('Lmplot: Studytime và G3 với hue="sex"', fontsize=14)
plt.xlabel('Studytime (Giờ học mỗi tuần)', fontsize=12)
plt.ylabel('G3 (Điểm cuối kỳ)', fontsize=12)
plt.show()

# Định nghĩa legend_labels cho các giá trị của studytime
legend_labels = {1: '1-2 giờ', 2: '2-5 giờ', 3: '5-10 giờ', 4: '>10 giờ'}
# Vẽ boxplot: Phân phối G3 theo studytime với màu sắc tùy chỉnh
plt.figure(figsize=(8, 6))
sns.boxplot(x='studytime', y='G3', data=Student_Pandas, hue='studytime', palette='Pastel2')
plt.title('Boxplot: Phân phối G3 theo Studytime', fontsize=14)
# Cập nhật nhãn cho các giá trị của studytime
plt.xticks(ticks=[0, 1, 2, 3], labels=[legend_labels[i] for i in range(1, 5)], rotation=0)
handles, labels = plt.gca().get_legend_handles_labels()
new_labels = [legend_labels[int(label)] for label in labels]
plt.legend(handles, new_labels, title='Studytime', title_fontsize='13', loc='lower center', fontsize='10')
plt.xlabel('Studytime (Giờ học mỗi tuần)', fontsize=12)
plt.ylabel('G3 (Điểm cuối kỳ)', fontsize=12)
plt.show()

# Vẽ pairplot: Mối quan hệ giữa G1, G2, G3, studytime
g = sns.pairplot(filtered_data, hue='studytime', palette='Set1')
plt.suptitle('Pairplot: Mối quan hệ giữa G1, G2, G3 và Studytime', fontsize=14, y=1.02)
handles, labels = g._legend_data.values(), g._legend_data.keys()
new_labels = [legend_labels[int(label)] for label in labels]
g._legend.remove()
plt.legend(handles, new_labels, title='Studytime', title_fontsize='13', 
           loc='center left', bbox_to_anchor=(1, 1), fontsize='10')
plt.tight_layout()  # Giúp tránh bị trùng lặp layout khi đặt legend ngoài
plt.show()
```

```python
# Vẽ Pie Chart theo Studytime 4 khung giờ học khác nhau
studytime_counts = Student_Pandas['studytime'].value_counts()
# colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
colors = ['red', 'green', 'blue', 'yellow']
labels = ['1-2 giờ', '2-5 giờ', '5-10 giờ', '>10 giờ']
plt.figure(figsize=(8, 6))
plt.pie(studytime_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Biểu đồ Pie Chart: Phân bố Studytime của sinh viên', fontsize=14)
plt.axis('equal')  # Đảm bảo biểu đồ tròn
plt.show()
```

```python
# Vẽ Correlation Heatmap theo điểm số cuối kỳ G3
plt.figure(figsize=(10, 8))
# Chỉ chọn các cột số để tính correlation
numeric_columns = Student_Pandas.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True, linewidths=0.5)
plt.title('Correlation Heatmap của các biến số trong Student_Pandas', fontsize=16)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()  # Giúp tránh bị trùng lặp layout khi đặt legend ngoài
plt.show()
```

```python
# vẽ subplot (Line graph, Bar chart, Scatter plot) theo HR_comma_sep.csv
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Đọc dữ liệu HR
df_hr = pd.read_csv('HR_comma_sep.csv')

# 1. Line Chart: Mối quan hệ giữa satisfaction_level và last_evaluation
axs[0].plot(df_hr.index[:100], df_hr['satisfaction_level'][:100], 'b-', label='Satisfaction Level', linewidth=2)
axs[0].plot(df_hr.index[:100], df_hr['last_evaluation'][:100], 'r--', label='Last Evaluation', linewidth=2)

# Tìm max và min cho Satisfaction Level
satisfaction_data = df_hr['satisfaction_level'][:100]
max_idx_sat = satisfaction_data.idxmax()
min_idx_sat = satisfaction_data.idxmin()
max_val_sat = satisfaction_data.max()
min_val_sat = satisfaction_data.min()

# Tìm max cho Last Evaluation
evaluation_data = df_hr['last_evaluation'][:100]
max_idx_eval = evaluation_data.idxmax()
max_val_eval = evaluation_data.max()

# Thêm annotations cho Min của Satisfaction Level
axs[0].annotate(f'Min Satisfaction\n({min_idx_sat}, {min_val_sat:.2f})', 
                xy=(min_idx_sat, min_val_sat), xytext=(min_idx_sat+10, min_val_sat+0.15),
                arrowprops=dict(facecolor='darkblue', shrink=0.05, alpha=0.7),
                fontsize=9, color='darkblue', ha='center')

# Thêm annotation cho Max của Last Evaluation
axs[0].annotate(f'Max Evaluation\n({max_idx_eval}, {max_val_eval:.2f})', 
                xy=(max_idx_eval, max_val_eval), xytext=(max_idx_eval+15, max_val_eval-0.1),
                arrowprops=dict(facecolor='red', shrink=0.05, alpha=0.7),
                fontsize=9, color='red', ha='center')

axs[0].set_title('Line Chart: Satisfaction Level vs Last Evaluation (100 nhân viên đầu)', fontsize=12, fontweight='bold')
axs[0].set_xlabel('Index nhân viên')
axs[0].set_ylabel('Điểm số (0-1)')
axs[0].legend()
axs[0].grid(True, alpha=0.3)

# 2. Bar Chart: Số lượng nhân viên theo mức lương
salary_counts = df_hr['salary'].value_counts()
bars = axs[1].bar(salary_counts.index, salary_counts.values, 
                  color=['red', 'orange', 'green'], 
                  edgecolor='black', alpha=0.8)
axs[1].set_title('Bar Chart: Số lượng nhân viên theo mức lương', fontsize=12, fontweight='bold')
axs[1].set_xlabel('Mức lương')
axs[1].set_ylabel('Số lượng nhân viên')
# Thêm giá trị trên mỗi cột
for bar in bars:
    height = bar.get_height()
    axs[1].text(bar.get_x() + bar.get_width()/2., height,
                f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# 3. Scatter Plot: Mối quan hệ giữa satisfaction_level và number_project
scatter = axs[2].scatter(df_hr['satisfaction_level'], df_hr['number_project'], 
                        c=df_hr['left'], cmap='coolwarm', alpha=0.6, s=30)
axs[2].set_title('Scatter Plot: Satisfaction Level vs Number of Projects', fontsize=12, fontweight='bold')
axs[2].set_xlabel('Satisfaction Level')
axs[2].set_ylabel('Number of Projects')
# Thêm colorbar để giải thích màu sắc
cbar = plt.colorbar(scatter, ax=axs[2])
cbar.set_label('Left (0=Stayed, 1=Left)', rotation=270, labelpad=15)

# Điều chỉnh khoảng cách giữa các subplot
plt.tight_layout(pad=3.0)
plt.show()
```

```python
# from scipy.stats import gaussian_kde
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot Example')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values     = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Ví dụ về biểu đồ radar')
plt.show()
```
