```python
import matplotlib.pyplot as plt
```

```python
# Vẽ biểu đồ đường Line Chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

plt.plot(x,y, 'bo-', label='Dữ liệu') #Vẽ đường 1
# plt.plot(x1,y1, 'ro-', label='Dữ liệu 1')
plt.plot(x1,y1, marker='o', color='r', linestyle='-', label='Dữ liệu') #Vẽ đường 2
plt.legend() #Hiển thị chú giải
plt.xlabel('Trục x') #Đặt nhãn cho trục x
plt.ylabel('Trục y') #Đặt nhãn cho trục y
plt.show()
```

```python
# Vẽ biểu đồ cột Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color='g', alpha=0.6)
plt.xlabel('Danh mục') #Đặt nhãn cho trục x
plt.ylabel('Giá trị') #Đặt nhãn cho trục y
plt.show()
```

```python
# Vẽ biểu đò hình tròn Pie Chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10] # tính theo %
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels = my_labels, colors = my_colors, autopct = '%1.1f%%')
plt.title('Biểu đò hình tròn')
plt.show()
```

```python
# Vẽ biểu đồ phân tán Scatter Plot
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')
plt.title('Ví dụ về biểu đồ phân tán')
plt.legend() #Hiển thị chú giải
plt.xlabel('Trục x') #Đặt nhãn cho trục x
plt.ylabel('Trục y') #Đặt nhãn cho trục y
plt.show()
```

```python
# Vẽ biểu đồ hộp Box Plot
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]
      
plt.boxplot(data)
plt.title('Ví dụ về biểu đồ hộp')
plt.ylabel('Giá trị') #Đặt nhãn cho trục y
plt.show()
```

```python
# Vẽ biểu đồ Violin plot
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]
      
plt.violinplot(data)
plt.title('Ví dụ về biểu đồ Violin')
plt.ylabel('Giá trị') #Đặt nhãn cho trục y
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

my_sentence = "Python is an amazing language for data visualization and word clouds."

wordcloud = WordCloud(width=800, height=400).generate(my_sentence)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Ví dụ về biểu đồ đám mây từ văn bản')
plt.show()

# To save the word cloud as an image:
# wordcloud.to_file("my_wordcloud.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values     = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Ví dụ về biểu đồ radar')
plt.show()
```

```python
#Ví dụ: Heatmap với dữ liệu tương quan (correlation matrix)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Tạo DataFrame mẫu
data = {
    'Math': [85, 90, 78, 92, 88],
    'Physics': [80, 85, 75, 89, 84],
    'Chemistry': [83, 87, 77, 90, 86],
    'English': [70, 72, 68, 75, 73]
}
df = pd.DataFrame(data, index=['Student A', 'Student B', 'Student C', 'Student D', 'Student E'])

# Tính hệ số tương quan giữa các môn
corr = df.corr()

# Vẽ heatmap
plt.figure(figsize=(8, 6))
# sns.heatmap(corr, annot=True, cmap="YlGnBu", fmt=".2f") # YlGnBu là Yellow,Green,Blue
#có thể thay với mã màu khác: viridis / coolwarm / hot / magma hoặc Blues/Reds/Greens
sns.heatmap(corr, annot=True, cmap="YlGnBu", fmt=".0%") # Hiển thị theo định dạng phần trăm
plt.title("Heatmap - Correlation Between Subjects")
plt.show()
```

```python
# Ví dụ khác: Heatmap trực tiếp từ ma trận số
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Ma trận số ngẫu nhiên 5x5
matrix = np.random.rand(5, 5)

plt.figure(figsize=(6, 5))
sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt=".2f") # Hiển thị 2 chữ số thập phân
plt.title("Random Heatmap")
plt.show()
```
